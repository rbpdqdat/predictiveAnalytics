```python
no_cols
```

```python
#find number of rows
counter = 0

main_dict = {}
for col in cols:
    main_dict[col]=[]
    
```

```python
#
for line in data:
    values = line.strip().split(',')
    for i in range(len(cols)):
        main_dict[cols[i]].append(values[i])
    counter += 1  
                
```
